```python
# resources taken from https://catalog.ourworldindata.org/
import os
from owid import catalog #API from the ourworldindata
import numpy as numpy
import pandas as pd
```

```python
#fetching the data from the api and url
rc = catalog.RemoteCatalog()
uri = "garden/covid/latest/cases_deaths/cases_deaths"
df_cases_deaths = rc[uri]
```

```python
# Get the first five rows
df_cases_deaths.head(5)
```

```python
df_cases_deaths.reset_index(inplace=True)
```

```python
df_cases_deaths['country'] = df_cases_deaths['country'].astype(str)
```

```python
print(df_cases_deaths.columns)
```

```python
#making all null objects to 0 because not all of the data/columns started at the same time
df_cases_deaths.fillna(0, inplace=True)
```

```python
# Get the data types and number of non-null values in the dataset.
df_cases_deaths.info()
```

```python
df_cases_deaths.head(5)
```

```python
#showing the last five rows
df_cases_deaths.tail(5)
```

```python
df_cases_deaths.to_csv('cases_deaths.csv', index=False)
```

```python
uri = "garden/covid/latest/testing/testing"
df_testing = rc[uri]
```

```python
df_testing.head(5)
```

```python
df_testing.reset_index(inplace=True)
```

drop most of the columns because they are rolling sums which are already set in stone and I also drop some on cases_deaths.csv on excel

```python
df_testing['country'] = df_testing['country'].astype(str)
```

```python
df_testing['date'] = pd.to_datetime(df_testing['date'].astype(str), errors='coerce')
```

```python
df_testing.info()
```

```python
df_testing.fillna(0, inplace=True)
```

```python
df_testing = df_testing.drop(columns=['total_tests', 'total_tests_per_thousand','new_tests_per_thousand','new_tests_7day_smoothed','new_tests_per_thousand_7day_smoothed'], axis=1)
```

```python
display(df_testing)
```

```python
df_testing.to_csv('tests_covid.csv', index=False)
```
